```python
import pandas as pd
import numpy as np
# Create a simple dataset
data = {
    "Name": ['Ajit Doval', 'Hamza Ali', None, 'Jassi Singh', 'Jameel Jamali', None], 
    "Marks": [90, 95, 92, 99, np.nan, np.nan] 
    }
df = pd.DataFrame(data)
# Replace 'NA' with np.nan
df.replace('NA', np.nan, inplace = True)
# Save the dataset
df.to_csv("data.csv", index  = False)
# Load it back
df = pd.read_csv("data.csv")
pd.set_option('display.max_rows', None)
# access specific columns
df.Name
df[['Name', 'Marks']]
# see all the column names
df.columns
# access rows by integer location
df.iloc[0]
df.iloc[[0, 1]]
# access rows and columns using integer location- (row, column)
df.iloc[[0, 2], [0,1]]
# using loc
df.loc[[0, 3], 'Name']
df.loc[[0, 3], ['Name', 'Marks']]
# Filter marks
df[df["Marks"] > 90]
# reverse selection
df.iloc[::-1]
# select last 3 rows
df.iloc[-3:]
# multi-condition filter
df["Honor roll"] = df["Marks"] >= 95
df.loc[(df["Marks"] > 90) & (df["Honor roll"] == True)]
# setting the index
df.set_index("Name", inplace = True)
df.index
df.loc["Ajit Doval", 'Marks']
df.reset_index(inplace = True)
# Anyone who has 'J' in their name
df[df["Name"].str.contains('J')]
# add new rows
new_row = pd.DataFrame([{'Name': 'Pinda Yaara', 'Marks': 57}])
df = pd.concat([df, new_row], ignore_index=True)
# Create new columns
df["Grade"] = df["Marks"] / 25
df.loc[df["Marks"] >= 95, "Status"] = "Excellent"
df.loc[df["Marks"] < 95, "Status"] = "Good"
df.loc[df["Honor roll"] == True, "PerformanceScore"] = df["Marks"] * 1
df.loc[df["Honor roll"] == False, "PerformanceScore"] = df["Marks"] * 0.9
# Rename the column
df.rename(columns = {"Grade": "GPA"}, inplace = True)
df
# Runnung "apply" on a series
df["Name"].apply(len)
def update_name(name):
    return name.upper()
df["Name"].apply(update_name)
df["Name"] = df["Name"].apply(lambda x : x.upper())
# running "apply" on a dataframe
df.apply(pd.Series.min)
# running "map"
df["Name"].map(len)
df["Name"].map(str.lower)
df["Name"].map({"Ajit" : "Pinda", "Hamza" : "Alam"})
# using "replace": wont put Null for values not assigned
df["Name"].replace({"Ajit" : "Pinda", "Hamza" : "Alam"})
# Removing the Status column
df.drop(columns = "Status", inplace = True)
# add new columns 
df[["first_name", "last_name"]] = df["Name"].str.split(" ", expand = True)
# Removing a row
df.drop(index = 3)
# Sorting by last name
df = df.sort_values(by = "last_name", ascending = False)
# Handling missing values
df.dropna(axis = 'index', how = 'all', subset = ["Name"])
# Check for missing values
df.isna()
# Fill missing values with a specific value
df["Marks"].fillna(df["Marks"].mean(), inplace = True)
# Convert Marks to float
df["Marks"] = df["Marks"].astype(float)
# Check data types
df.dtypes
```
